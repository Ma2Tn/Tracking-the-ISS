```python
import requests
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
times= []
long= []
lat= []



for i in range (0,120):
    a= requests.get("http://api.open-notify.org/iss-now.json")
    long.append(a.json()['iss_position']['longitude'])
    lat.append(a.json()['iss_position']['latitude'])
    times.append(a.json()['timestamp'])
    time.sleep(10)

print(long)
print(lat)
print(times)
```

```python
long2 = [float(i) for i in long]
lat2 = [float(i) for i in lat]
x = np.array(long2)
y = np.array(lat2)


plt.title("ISS Location at Certain Timepoint")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.scatter(x, y)

plt.xticks(np.arange(min(long2), max(long2), 10))
plt.yticks(np.arange(min(lat2), max(lat2), 5))



plt.show()
```
